```python
#Splitting df_train into X_train and Y_train and scaling X_train and X_test:
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
X_train = std_scaler.fit_transform(df_train.drop('Default', axis=1))
Y_train = df_train['Default']

X_test = std_scaler.fit_transform(df_test)
```

```python
#Fitting Logistic Regression to the training data:
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter_grid = {'penalty' : ['l1', 'l2'], 'C' : [0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(estimator = log_reg, param_grid = parameter_grid, cv=4, verbose=0)
best_model = grid.fit(X_train, Y_train)
```

```python
print('Best Penalty:', best_model.best_estimator_.penalty)
print('Best C:', best_model.best_estimator_.C)
print(best_model.best_estimator_)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty='l2', C=0.001)

from sklearn.feature_selection import RFE
rfe = RFE(log_reg, 18)
rfe = rfe.fit(X_train, Y_train)
# summarize the selection of the attributes
print(rfe.support_)
print(rfe.ranking_)
```

```python
#Predicting default values for test data:
rfe.predict(df_test)
```

```python
# Converting array to dataframe column:
columns = ['Default_test']
values = rfe.predict(X_test)
predictions = pd.DataFrame(values, columns = columns)
```
